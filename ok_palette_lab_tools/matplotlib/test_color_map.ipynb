```python
import numpy
import matplotlib.colors
import matplotlib.pyplot
import ok_palette_lab.plotly
from ok_palette_lab_tools.test_data.gaussian import generate_gaussian_data
```

```python
x, y, z = generate_gaussian_data()
```

```python
color_map = matplotlib.colors.LinearSegmentedColormap.from_list(
    name="autumn", colors=ok_palette_lab.plotly.autumn
)
```

```python
matplotlib.pyplot.figure()
matplotlib.pyplot.imshow(
    z, extent=(x.min(), x.max(), y.min(), y.max()), origin="lower", cmap=color_map
)
matplotlib.pyplot.colorbar()
matplotlib.pyplot.show()
```
